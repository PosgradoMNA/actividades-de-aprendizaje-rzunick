```python
import pandas as pd
import numpy as np

x = {"Company" : ["Ford","Ford","VW","BMW","Cooper","Cooper"], 
     "Stars" : [1, 2, np.nan, 2, 1, 1],
     "Weight" : [2,4,2,2,3, None],
     "Origin" : ["Mexico", "China", "Mexico", None, "China", np.nan],
     "Length" : [50, 40, 30, np.nan, 45, pd.NaT]
}
```

```python
df2 = pd.DataFrame(data = x)
df2
```

```python
#Solución 1 - Eliminar todos los renglones, si tienen al menos un dato perdido
ndf2 = df2.dropna()
ndf2
```

```python
#Solución 2 - Eliminar columna, si tienen al menos un dato perdido en una columna.
ndf = df2.copy()
ndf.dropna(inplace = True,
           axis = 1
           )
ndf
```

```python
#Solución 3 - Eliminar los renglones si todos los valores son perdidos
ndf = df2.copy()
ndf.dropna(how = "all", 
           inplace = True)
ndf
```

```python
#Solución 4 - Si existe al menos 3 datos, no se eliminará el renglón
ndf = df2.copy()
ndf.dropna(thresh =3, 
           inplace = True,
           )
ndf
```

```python
#Substituir NA con mean
ndf = df2.copy()
ndf["Stars"].fillna(value = ndf["Stars"].mean(),
                    inplace = True
                    )
ndf
```

```python
ndf = df2.copy()
sm = ndf["Stars"].median()

ndf["Stars"].fillna(value = sm,
                    inplace = True
                    )
ndf
```

```python
ndf["Origin"].mode()
```

```python
oc = ndf["Origin"].mode()[0]
```

```python
#Columna Origin se limpian datos perdidos utilizando moda. 
ndf["Origin"].fillna(value = oc,
           inplace = True
           )
ndf
```
